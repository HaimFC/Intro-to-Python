Write a function that gets a sorted list of numbers and another number X. The
function decides whether or not the given number X is inside the list or not and
returns True/False answer.

```python
def binary_search(sorted_list, x):
    # ---------Classic Binary search-------------
    left, right = 0, len(sorted_list) - 1
    while left <= right:
        mid = (left + right) // 2
        if sorted_list[mid] == x:
            return True
        elif sorted_list[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return False

def main():
    sorted_list = [1, 3, 5, 7, 9, 11, 13, 15]
    x = 7
    result = binary_search(sorted_list, x)
    print(result)

if __name__ == "__main__":
    main()
```
