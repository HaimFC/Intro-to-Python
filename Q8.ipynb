Given two txt files that have lists of numbers in them, find the numbers that are
overlapping.
You should generate the txt files by yourself (using python, not manually): One
txt file has a list of all prime numbers under 1000, and the other txt file has a list
of happy numbers up to 1000 .
[Prime numbers are numbers that canâ€™t be divided by any other number. Yes,
Happy numbers are a real thing in mathematics - you can look it up on
Wikipedia]

```python
def Q2(number):
  divisors = [i for i in range(1, number + 1) if number % i == 0]
  return divisors

def Q4(number):
  divisors = Q2(number)
  if len(divisors) == 2:
    return True
  else:
    return False

def is_happy(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit)**2 for digit in str(n))
    return n == 1

def generate_primes(filename, limit=1000):
    primes = [num for num in range(2, limit) if Q4(num)]
    with open(filename, 'w') as file:
        file.write('\n'.join(map(str, primes)))

def generate_happy_numbers(filename, limit=1000):
    happy_numbers = [num for num in range(1, limit + 1) if is_happy(num)]
    with open(filename, 'w') as file:
        file.write('\n'.join(map(str, happy_numbers)))

def find_overlapping_numbers(file1, file2):
    with open(file1, 'r') as f1, open(file2, 'r') as f2:
        set1 = set(map(int, f1.readlines()))
        set2 = set(map(int, f2.readlines()))
    overlap = sorted(set1 & set2)
    return overlap

def main():
    primes_file = "primes.txt"
    happy_numbers_file = "happys.txt"
    generate_primes(primes_file)
    generate_happy_numbers(happy_numbers_file)
    overlap = find_overlapping_numbers(primes_file, happy_numbers_file)
    print(f"Overlapping numbers: {overlap}")

if __name__ == "__main__":
    main()
```
