Use the BeautifulSoup library and requests Python packages to print out a list
of all the article titles on the New York Times / Ynet homepage

```python
import requests
from bs4 import BeautifulSoup

def getArticles(url):
    articles = []
    if url == 'https://www.nytimes.com':
        rss_url = 'https://rss.nytimes.com/services/xml/rss/nyt/HomePage.xml'
        response = requests.get(rss_url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'xml')
            items = soup.find_all('item')
            for item in items:
                title = item.title.get_text(strip=True)
                articles.append(title)
        else:
            print("Error fetching NYTimes RSS feed")
            return None
    elif url == 'https://www.ynet.co.il':
        response = requests.get(url)
        response.encoding = 'utf-8'
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            title_tags = ['h1', 'h2', 'h3', 'a']
            for tag in title_tags:
                for element in soup.find_all(tag):
                    title = element.get_text(strip=True)
                    if title and len(title) > 10:
                        articles.append(title)
        else:
            print("Error fetching Ynet homepage")
            return None
    return articles

def main():
    # Ynet Articles
    ynet_url = 'https://www.ynet.co.il'
    ynet_articles = getArticles(ynet_url)
    print("Ynet Articles:")
    if ynet_articles:
        for article in ynet_articles:
            print(article)

    # NYTimes Articles
    nytimes_url = 'https://www.nytimes.com'
    nytimes_articles = getArticles(nytimes_url)
    print("\nNYTimes Articles:")
    if nytimes_articles:
        for article in nytimes_articles:
            print(article)

if __name__ == "__main__":
    main()
```
